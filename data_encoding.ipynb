```python
df = pd.DataFrame({"name":["wscube","cow","cat","dog","black"]})
df
```

**ordinal encoding**

```python
df1 = pd.DataFrame({"size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df1
```

```python
df1 = pd.DataFrame({"size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
ord_ds = [["s","m","l","xl"]]
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = ord_ds)
df1["size"] = oe.fit_transform(df1[["size"]])
df1
```
