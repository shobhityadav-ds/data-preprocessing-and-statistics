```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("/content/drive/MyDrive/Outlier_Loan_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
sns.distplot(df['Current_Balance'])
plt.title('Current_Balance')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['CurrentBalance_ss']= pd.DataFrame(scaler.fit_transform(df[['Current_Balance']]),columns=["x"])
df.head()
```

```python
df.describe()
```

```python
plt.subplot(1,2,1)
sns.distplot(df['CurrentBalance_ss'],color="red")
plt.title('CurrentBalance_ss')
plt.subplot(1,2,2)
sns.distplot(df['Current_Balance'],color = "green")
plt.title('Current_Balance')
plt.show()
```

**normalization**

```python
from sklearn.preprocessing import MinMaxScaler
ss = MinMaxScaler()
df['CurrentBalance_minmax']= pd.DataFrame(ss.fit_transform(df[['Current_Balance']]),columns=["x"])
df.head()
```

```python
plt.subplot(1,2,1)
sns.distplot(df['CurrentBalance_minmax'],color="red")
plt.title('CurrentBalance_normal')
plt.subplot(1,2,2)
sns.distplot(df['Current_Balance'],color = "green")
plt.title('Current_Balance')
plt.show()
```
