```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
import seaborn as sns
ds = pd.read_csv('/content/drive/MyDrive/diabetes.csv')
ds.head()
```

**forward elimination**

```python
ds.isnull().sum()
```

```python
x = ds.iloc[:,:-1]
y = ds['Outcome']
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
fs = sfs(lr,k_features=1,forward=True)
fs = fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```

**backward elimination**

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
fs = sfs(lr,k_features=4,forward=False)
fs = fs.fit(x,y)
```
